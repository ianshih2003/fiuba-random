```
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```
dataset = pd.read_csv("properati_argentina_2021.csv")
```

```
cond_property_types = dataset['property_type'].isin(['Casa', 'PH', 'Departamento'])
cond_operation = dataset['operation'] == 'Venta'
cond_price_currency = dataset['property_currency'] == 'USD'
cond_location = dataset['place_l2'] == 'Capital Federal'

dataset = dataset[cond_property_types & cond_operation & cond_price_currency & cond_location]
```

```
ds_cropped = dataset.drop(['id', 'property_currency', 'place_l2', 'operation'], axis='columns')
ds_cropped.columns.tolist()
```

```
copy_ds_cropped = ds_cropped.copy()
copy_ds_cropped = copy_ds_cropped[['property_rooms', 'property_bedrooms']]

#z-score property_rooms
media_property_rooms=np.mean(copy_ds_cropped.property_rooms)
std_property_rooms=np.std(copy_ds_cropped.property_rooms)
copy_ds_cropped['z_property_rooms']=(copy_ds_cropped.property_rooms - media_property_rooms)/std_property_rooms

#z-score property_bedrooms
media_property_bedrooms=np.mean(copy_ds_cropped.property_bedrooms)
std_property_bedrooms=np.std(copy_ds_cropped.property_bedrooms)
copy_ds_cropped['z_property_bedrooms']=(copy_ds_cropped.property_bedrooms - media_property_bedrooms)/std_property_bedrooms
```

```
#z-score modificado property_rooms
mediana_property_rooms=np.median(copy_ds_cropped.property_rooms)
MAD=np.median(np.absolute(copy_ds_cropped.property_rooms-mediana_property_rooms))

copy_ds_cropped['zm_property_rooms']=(copy_ds_cropped.property_rooms - mediana_property_rooms)*0.6745/MAD

#z-score modificado property_bedrooms
mediana_property_bedrooms=np.median(copy_ds_cropped.property_bedrooms)
MAD=np.median(np.absolute(copy_ds_cropped.property_bedrooms-mediana_property_bedrooms))

copy_ds_cropped['zm_property_bedrooms']=(copy_ds_cropped.property_bedrooms - mediana_property_bedrooms)*0.6745/MAD
```

```
#Calculo el vector de medias
vmedias=np.mean(copy_ds_cropped[['property_rooms','property_bedrooms']])

#Calculo la diferencia entre las observaciones y el vector de medias
x_dif = copy_ds_cropped[['property_rooms','property_bedrooms']] - vmedias

#Calculo matriz de covarianza y su inversa
cov=copy_ds_cropped[['property_rooms','property_bedrooms']].cov().values
inv_cov = np.linalg.inv(cov)

#Calculamos el cuadrado de la distancia de mahalanobis
mahal =np.dot( np.dot(x_dif, inv_cov) , x_dif.T)

copy_ds_cropped['mahalanobis']=mahal.diagonal()# Start writing code here...
```
